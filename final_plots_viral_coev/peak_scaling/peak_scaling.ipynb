```julia
using DataFrames
using Plots
using Statistics
using CSV
```

```julia
df = CSV.read("survival_probabilities.csv", DataFrame)
const MIGRATION_RATES = exp10.(LinRange(-6.0, -2, 30))
const HOST_POPULATION_SIZES = exp10.(LinRange(4.0,7.0,7))
```

```julia
p = plot()
opt_migration = []
for k = 1:7
    y_data = df[df.HostPerDemeIdx .== k, "SurvivalProbability"]
    scatter!(p, y_data, label=log10(HOST_POPULATION_SIZES[k]))
    push!(opt_migration, MIGRATION_RATES[argmax(y_data)])
end
display(p)
```

```julia
p = scatter(HOST_POPULATION_SIZES, opt_migration, yscale=:log10, xscale=:log10, xlabel="host population size", ylabel="peak migration rate")
plot!(p, HOST_POPULATION_SIZES, 10^-2 ./ (HOST_POPULATION_SIZES).^0.5, linestyle=:dash)
display(p)
```
