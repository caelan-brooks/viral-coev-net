```julia
using DataFrames
using Plots
using Statistics
using CSV
```

```julia
df = CSV.read("survival_probabilities.csv", DataFrame)
```

```julia
p = plot()
opt_migration = []
for k = 1:7
    y_data = df[df.HostPerDemeIdx .== k, "SurvivalProbability"]
    scatter!(p, y_data)
    push!(opt_migration, argmax(y_data))
end
display(p)
```

```julia
p = scatter(opt_migration[1:6])
x_vals = 1:6
plot!(p, x_vals, -2 * x_vals .+ 17, linestyle=:dash)
display(p)
```
