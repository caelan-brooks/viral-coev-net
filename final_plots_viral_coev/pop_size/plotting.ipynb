```julia
using CSV
using DataFrames
using Plots

# Load the data
data = CSV.read("analysis_results.csv", DataFrame)

# Extract the relevant columns
host_population_size = data.HostPopulationSize
survival_probability = data.SurvivalProbability

# Create the plot
scatter(host_population_size, survival_probability,
    xscale=:log10,
    ylims=(-0.1,1.1),
    xlabel="Host Population Size", 
    ylabel="Survival Probability",
    # title="\\textbf{Survival Probability vs Host Population Size}",
    lw=3, # thick line
    legend=:none,
    fontfamily="Computer Modern", # LaTeX fonts
    tickfont=font(12)
    )
```

```julia

# Save or display the plot
savefig("survival_probability_vs_host_population_size.png") # To save the figure
```
